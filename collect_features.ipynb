```python
### merge with baits information

geneLeast = 50
datadir = '/Users/Yuan/Documents/BLab/Predict_target_genes/hMM/Features_data_Lease%d' % geneLeast


Fs = [t.replace('.bed','') for t in [f for f in os.listdir('TFBS/')] if 'bed' in t] +  \
     [t.replace('.bed','') for t in [f for f in os.listdir('HM/')] if 'bed' in t] +  \
     ['DNase_merged']


    
### read in the data

files = [f for f in os.listdir('%s/features' % datadir)  if 'matrix.bed' in f]

samples = []
for f in files:
    fn = open('%s/features/%s' % (datadir,f),'r')
    fn.readline()
    for line in fn:
        samples.append([f.replace('_matrix.bed','')] + line.rstrip().split('\t')[1:])
    fn.close()
    
    
sample_df = pd.DataFrame(samples)
sample_df.index = sample_df[0]
del sample_df[0]

fn = open('%s/features/%s' % (datadir,f),'r')
sample_df.columns = np.array(fn.readline().rstrip().split('\t')[1:])
fn.close()

sample_df = sample_df.astype('float')


### Do the merging

sample_df['for_merge'] = ['_'.join((x[0], str(int(x[1])),str(int(x[2])))) for x in zip(sample_df.index, sample_df['start'], sample_df['end'])]

capture_hic_data = pd.read_csv('capture_hic/PCHiC_peak_matrix_cutoff5.txt',sep='\t')
capture_hic_data = capture_hic_data[capture_hic_data['tCD4']>0]
capture_hic_data.index = [str(x).replace('.','').replace(';','_') for x in np.array(capture_hic_data['baitName'])]
capture_hic_baits = capture_hic_data[['baitStart','baitEnd', 'oeStart', 'oeEnd']]
capture_hic_baits['for_merge'] = ['_'.join((x[0], str(int(x[1])),str(int(x[2])))) for x in zip(capture_hic_baits.index, capture_hic_baits['oeStart'], capture_hic_baits['oeEnd'])]
capture_hic_baits = capture_hic_baits.drop_duplicates()
capture_hic_baits = capture_hic_baits[~capture_hic_baits['for_merge'].duplicated()]

# perserve the index when merge with another dataframe
del sample_df.index.name
data = sample_df.reset_index().merge(capture_hic_baits, left_on='for_merge', right_on = 'for_merge', how="left").set_index('index')


bait_location = np.array(np.mean(data[['baitStart','baitEnd']], axis=1))
oe_location = np.array(np.mean(data[['oeStart','oeEnd']], axis=1))
data['Distance'] = np.abs(oe_location - bait_location)



### save

data = data[['Distance'] + ['tCD4'] + Fs]
data.to_csv('%s/Features.csv' % datadir,sep='\t')


### plot the raw features

plt.figure()
X = np.array(data)
X = np.ma.masked_where(np.abs(X) < 1e-3, X)
cmap = mpl.cm.jet
cmap.set_bad('white')
cax = plt.imshow(X, interpolation='nearest', cmap=cmap, aspect='auto')
plt.colorbar(cax)
plt.savefig('%s/Feaures_Raw.png' % datadir)
plt.close()







#### Normazlie the data

##https://github.com/ShawnLYU/Quantile_Normalize/blob/master/quantile_norm.py

def quantileNormalize(df_input):
    df = df_input.copy()
    #compute rank
    dic = {}
    for col in df:
        dic.update({col : sorted(df[col])})
    sorted_df = pd.DataFrame(dic)
    rank = sorted_df.mean(axis = 1).tolist()
    #sort
    for col in df:
        t = np.searchsorted(np.sort(df[col]), df[col])
        df[col] = [rank[i] for i in t]
    return df


data_normalized = quantileNormalize(data)
data_normalized = np.log2(data_normalized + 1e-10)
data_normalized = data_normalized.apply(lambda x: (x-np.mean(x)) / np.std(x), axis=0)


data_normalized.to_csv('%s/Features_normalized_Least%d_interactions.csv' % (datadir, geneLeast),sep='\t')



### plot the raw features

plt.figure()
X = np.array(data_normalized)
X = np.ma.masked_where(np.abs(X) < 1e-3, X)
cmap = mpl.cm.jet
cmap.set_bad('white')
cax = plt.imshow(X, interpolation='nearest', cmap=cmap, aspect='auto')
plt.colorbar(cax)
plt.savefig('%s/Feaures_normalized.png' % datadir)
plt.close()

```
